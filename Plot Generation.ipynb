```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def generateLineplot(x, uniform_y, misplaced_y, manhattan_y, x_label, y_label, title):

    fig, ax = plt.subplots()
    
    # Plot a simple line chart
    ax.plot(x, uniform_y , label='Uniform Cost Search')
    ax.plot(x, misplaced_y, label='A* Search with Misplaced Tiles Heuristic')
    ax.plot(x, manhattan_y, label='A* Search with Manhattan Distance Heuristic')

    ax.set_xlabel(x_label, fontweight ='bold', fontsize = 15)
    ax.set_ylabel(y_label, fontweight ='bold', fontsize = 15)
    
    plt.legend()
    plt.title(title)
    #plt.savefig(title+".png")
    plt.show()
    
```

```python
# node expand vs depth
x = [0, 2, 4, 8, 12, 16, 20, 24] 

uniform_y = [0, 5, 25, 233, 1775, 12379, 59346, 234401]
misplaced_y = [0, 2, 4, 16, 93, 505, 2437, 14530]
manhattan_y = [0, 2, 4, 12, 28, 56, 220, 1539]

generateLineplot(x, uniform_y, misplaced_y, manhattan_y, 
                 'Depth', 'Node Expanded', 'Depth vs Node Expanded')
```
